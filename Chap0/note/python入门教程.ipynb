## python的一些基础概念


### 数据类型

* **Numbers** （数字）
   在powershell的交互环境里，可以通过命令语句实现计算器等简单的功能

```
PS C:\Users\admin\test> python
Python 3.6.2 (v3.6.2:5fd33b5, Jul  8 2017, 04:57:36) [MSC v.1900 64 bit (AMD64)] on win32
Type "help", "copyright", "credits" or "license" for more information.
>>> 2+3 
5
>>> 2+3*2 
8
>>> 17/3 
5.666666666666667
>>> round(_,3) #控制小数的精度，区3位小数
5.667
>>> 17//3 # 取整数的写法，也可做int(17/3),int()表示转换为整数类型
5
>>> 17%3
2
>>>2**3
8
```
四则运算可直接输入，回车得到结果。整数的数据类型是int，含小数位的数据类型是float，幂函数用x**y表示。


* **Strings**（字符串）
   * 字符串一般用双引号或单引号括起来。输出带引号。当字符串内出现单引号时，外围的用双引号区分，或者用转义字符表示。当出现更复杂的情况的时候，在交互模式下也需要用到print函数。
 ```
 >>> 'apples'
'apples'
>>> "don't"
"don't"
>>> 'don\'t'
"don't"
 ```
   * 字符串可通过 + 号实现字符融合(字符与字符之间可直接放在一起)，*号实现字符重复
 ```  
>>> 'i''am'
'iam'
>>> 'py'+'thon'+'yes'*2
'pythonyesyes'
 ```
   
   * 字符的索引提取。可通过String[position]的方式提取相应位置的字符
```
>>> word='python' #字符位置分别为0，1，2，3，4，5，倒数为-6，-5，-4，-3，-2，-1
>>> word[0]
'p'
>>> word[3]
'h'
>>> word[-1]
'n'
```

   * 字符串可索引进行切片分割。String[m:n]截取位置m到位置n（不包括n）的字符。

```
>>> word[2:4]
'th'
>>> word[:4]
'pyth'
>>> word[:42]
'python'
>>> word[3:]
'hon'
>>> word[:-1]
'pytho'

```

* **Lists**（列表）
  * 列表格式通常为squares=[a,b,c,d,e,f],可以是各种东西**数字**，**字符**甚至是**变量**，**自己定义的类**等。和字符串一样，列表可进行检索、切片和融合。但和字符串不同的是，字符串不可以更改，只能新建字符串，而列表是可以更改元素，包含的功能有：
     * list.append(x)列表后面增加一个元素
     * list.extend()在列表后面一次性追加另一个序列中的多个元素
     * list.insert(i,x)在列表的i位置插入一个元素
     * list.remove(x)删除列表里第一个为x的值
     * list.pop([i])弹出列表最后一个值，当有[i]位置时，弹出i位置的值
     * list.clear()清除列表里的所有值
     * list.index(obj)返回列表里第一个出现obj的索引位置
     * list.count(x)返回x出现的次数
     * list,sort(key=None,reverse=False)按一定条件排序...（待补充）
     * list.reverse()翻转列表，最后一个做第一个
     * list.copy()复制列表
     
 ```
>>> squares=[1,4,9,16,25]
>>> squares
[1, 4, 9, 16, 25]
>>> fruits=['apple', 'orange', 'banana', 'pear', 'kiwi','apple']
>>> fruits[0]
'apple'
>>> fruits[2:4]
['banana', 'pear']
>>> fruits.append('banana')
>>> fruits
['apple', 'orange', 'banana', 'pear', 'kiwi', 'apple', 'banana']
>>> fruits.pop()
'banana'
>>> fruits
['apple', 'orange', 'banana', 'pear', 'kiwi', 'apple']
>>> fruits.reverse()
>>> fruits
['apple', 'kiwi', 'pear', 'banana', 'orange', 'apple']
>>> fruits.count('apple')
2
>>> fruits.index('apple')
0
>>> fruits.sort()
>>> fruits
['apple', 'apple', 'banana', 'kiwi', 'orange', 'pear']
 
 ```

### 变量
* 程序中需要用到很多的数据，需要将它们用一个简单的名字代表，方便程序引用。变量就是代表某个数据（值）的名称。用等于号“=”一般用来给变量赋值，比如a=2,即把2赋值给a，变量可用来存放数值、字符串等。可以进行运算，相当于数学方程式里的x,y。需要先进行定义（赋初始值）才可使用，否则会返回错误。
* 变量的命名规范
 * 可以包括字母数字下划线，但是不可以数字开头
 * 系统关键字不能用来做变量名
 * 除了下划线，其他符号不能做变量名使用
 * 区分大小写
 
```
>>> a="hello"
>>> a
'hello'
>>> a=2
>>> a
2
>>> b
Traceback (most recent call last):
  File "<stdin>", line 1, in <module>
NameError: name 'b' is not defined
>>>b=3
>>>a+b
5
```

### if 语句
   
   if语句用于需要条件判断的情况。if后有要执行的语句，if行后应加冒号，下行应有缩进。
   
```
>>> x = int(input("Please enter an integer: ")) 
Please enter an integer: 24
>>> if x<0:
...     x=0
...     print('Negative changed to zero')
... elif x==0:
...     print('Zero')
... elif x==1:
...     print('single')
... else:
...     print('more')
...
more
    
```

### For循环

 * for循环可根据条件进行一直循环，用在程序里需要很多重复动作的情况。
 * for 循环的语法为：
   for 可迭代变量 in 数列:
       执行语句
  当条件变为假，循环结束
  
 ```
>>> fruits=['apple', 'orange', 'banana', 'pear', 'kiwi','apple']
>>> for f in fruits:
...     print(f,len(f))
...
apple 5
orange 6
banana 6
pear 4
kiwi 4
apple 5
>>>
 ```
 
 ### 函数
 
 * 函数相当于数学里的公式，能实现某个特定的功能，为使程序简化易读，将这些代码封装以便于随时调用。内置函数及为应用较多已封装好的程序，可用import ../from ..import.. 语句调用。
 * 用def命令创建一个函数，也就是difine的意思。紧挨着def的是函数的名字，名字可以随便取，但最好能体现函数的功能。括号里的是函数需要的值。紧接着是冒号，冒号下缩进四个空格的都是这个函数的内容。
   用户定义函数示例如下,将下列代码保存为test.py
   
 
 ```
 def p_two(*args): # *args表示把我们需要的量都放到名叫args的列表里去
    arg1, arg2 = args # 解包参数
    print("arg1:%r, arg2:%r" %(arg1,arg2)) # %r为格式化字符，即把右边变量的值放到%r所在的位置上
	
def p_one(arg1,arg2):
    print("arg1:%r, arg2:%r" %(arg1,arg2))
	
def one(arg1):
    print("arg1:%r"%(arg1))
	
def none():
    print("nothing")

p_two("i", "you")
p_one("i", "you")
one('i')
none()

 ```
 
 运行以上脚本，会看到以下结果：
 
 ```
 PS C:\Users\admin\test>python test.py
 arg1:'i', arg2:'you'
 arg1:'i', arg2:'you'
 arg1:'i'
 nothing
 ```
